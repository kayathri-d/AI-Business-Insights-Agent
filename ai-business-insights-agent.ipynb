# 🧠 Agent Architecture (Multi-Agent System)

This project is designed as a **multi-agent system** with three simple agents:

1. **Data Loader Agent**
   - Loads the sales dataset using `read_csv`.
   - Checks the data shape and first few rows.

2. **Insight Agent**
   - Calculates total sales, average monthly sales, best & worst months.
   - Computes region-wise sales and profit.
   - Prepares the final business summary.

3. **Reporting Agent**
   - Generates the text summary for the business user.
   - Creates the monthly sales trend chart for quick visualization.

The AI Business Insights Agent coordinates these agents to give clear, data-driven insights.

# 🔧 Tools Used by the Agent

The agent uses simple "tools" implemented as Python functions and libraries:

- **Dataset Tool:** `pd.read_csv()` to load the sales data.
- **Summary Tool:** Code that computes  
  - total sales  
  - average monthly sales  
  - best & worst month
- **Region & Profit Tool:** Code that groups by region to calculate sales and profit.
- **Visualization Tool:** `matplotlib` to plot the monthly sales trend.

These tools are combined by the agent to answer business questions automatically.

# Loading the Dataset📂

**⚡️This step loads the sales dataset into pandas DataFrame for analysis.**

```python
import pandas as pd

df = pd.read_csv('/kaggle/input/sample-salesdata/sample_sale.csv.txt')

df.head()
```

# Fetching Live Data using API

```python
import requests

# Currency Exchange API
url = "https://open.er-api.com/v6/latest/INR"

response = requests.get(url)
data = response.json()

usd_rate = data['rates']['USD']

print("💱 INR to USD Exchange Rate:", usd_rate)
```

# Business Calculation

```python
# Convert Total Sales from INR to USD

total_sales = df['Sales'].sum()
print(total_sales)

sales_usd = total_sales * usd_rate

print(f"💰 Total Sales in USD: {sales_usd:.2f}")
```

🌍 The AI Business Insights Agent also integrates real-time currency exchange data using APIs to support international business analysis.

# Basic Business Insights

**Basic Analysis:**

* 📊Total sales
* 📉Average sales
* 🏆Best & worst month

```python
#Calculating Basic business insights

total_sales = df['Sales'].sum()
average_sales = df['Sales'].mean()
best_month = df.loc[df['Sales'].idxmax(), 'Month']
worst_month = df.loc[df['Sales'].idxmin(), 'Month']

print("📊 Total Sales:", total_sales)
print(f"📈Average Monthly Sales: {average_sales:.2f}")
print("🥇Best Month:", best_month)
print("⚠️Worst Month:", worst_month)
```

# 🗂Agent Memory(Storing Key Insights)

```python
# Simple agent memory: store key insights
agent_memory = {
    "total_sales": total_sales,
    "average_monthly_sales": average_sales,
    "best_month": best_month,
    "worst_month": worst_month
}

print("🧾 Agent Memory (stored insights):")
for key, value in agent_memory.items():
    print(f"- {key}: {value}")
```

# 🧠 Agent Memory (Sessions & Context)

The agent keeps a small **memory dictionary** (`agent_memory`) that stores
important insights such as total sales, average monthly sales, best month, and
worst month. This simulates how an agent can remember key results within a
session and reuse them in later steps or responses.

## 📈 Advanced Business Insights (Trends & Regions)

**This section provides deeper insights such as monthly sales trends, region-wise performance, and profit analysis.**

```python
#Sales Trend by Month
print("📌 Sales Trend by Month:")
print(df[['Month', 'Sales']])

#Sales by Region
print("\n📌 Sales by Region:")
print(df.groupby('Region')['Sales'].sum())

#Profit by Region
print("\n📌 Profit by Region:")
print(df.groupby('Region')['Profit'].sum())

#Profit Insights
print("\n📌 Month with Highest Profit:", df.loc[df['Profit'].idxmax(), 'Month'])
print("📌 Month with Lowest Profit:", df.loc[df['Profit'].idxmin(), 'Month'])
```

# 🎯Final Business Summary

**This final summary combines all key insights from sales, regions, and profit analysis.**

```python
print("📊 BUSINESS SUMMARY\n")

print(f"✔ Total Sales: {total_sales}")
print(f"✔ Average Monthly Sales: {average_sales:.2f}")
print(f"✔ Best Month for Sales: {best_month}")
print(f"✔ Worst Month for Sales: {worst_month}")

#Region Performance
print("\n📌 Region Performance:")
region_sales = df.groupby('Region')['Sales'].sum()
for region, value in region_sales.items():
    print(f"   - Region {region}: {value} sales")

#Profit Insights
print("\n📌 Profit Insights:")
region_profit = df.groupby('Region')['Profit'].sum()
for region, value in region_profit.items():
    print(f"   - Region {region}: {value} profit")

print(f"\n🔥 Highest Profit Month: {df.loc[df['Profit'].idxmax(), 'Month']}")
print(f"⚠️ Lowest Profit Month: {df.loc[df['Profit'].idxmin(), 'Month']}")
```

# 📊Data Visualization

**This chart displays the monthly sales trend clearly.**

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(df['Month'], df['Sales'], marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
